```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('WFM_test.xlsx')
```

```python
data.head()
```

```python
#pd.to_datetime(data['Window End']) - pd.to_datetime(data['Window Start'])
data.shape
```

```python
hours = ['Productive Total Hours', 'Break Total Hours', 
         'Lunch Total Hours', 'Sick Total Hours']
data.sort_values(by=['Productive Total Hours'], ascending=False, inplace=True)
```

```python
f, ax= plt.subplots(figsize=(18,6))
data.plot(ax= ax, x='Agent Name', y = hours, kind='bar', stacked=True)
data.plot(ax=ax, x= 'Agent Name', y = 'Total Hours', label = 'Total Hours', color = 'black' )
plt.xticks(rotation=90)
plt.ylabel('Number of hours')
ax.set_title('Hours worked per agent');
```
